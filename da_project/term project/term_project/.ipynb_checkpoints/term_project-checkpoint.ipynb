as a reminder:

<b>Data Fields:</b><br/>
<b>datetime</b> - hourly date + timestamp  <br/>
<b>season</b> -  1 = spring, 2 = summer, 3 = fall, 4 = winter<br/>
<b>holiday</b> - whether the day is considered a holiday<br/>
<b>workingday</b> - whether the day is neither a weekend nor holiday<br/>
<b>weather</b> - 1: Clear, Few clouds, Partly cloudy, Partly cloudy<br/>
&emsp;&emsp;&emsp;&emsp;&emsp;2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist<br/>
&emsp;&emsp;&emsp;&emsp;&emsp;3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds<br/>
&emsp;&emsp;&emsp;&emsp;&emsp;4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog<br/>
<b>temp</b> - temperature in Celsius<br/>
<b>atemp</b> - "feels like" temperature in Celsius<br/>
<b>humidity</b> - relative humidity<br/>
<b>windspeed</b> - wind speed<br/>
<b>casual</b> - number of non-registered user rentals initiated<br/>
<b>registered</b> - number of registered user rentals initiated<br/>
<b>count</b> - number of total rentals<br/>

```python
#Before modeling, lets check what we finally have in both train.csv and test.csv files
#first on train.csv file
train_df.info()
#then on test.csv file
test_df.info()
```

To make both csv file consistent, we will add count in test.csv file and also will split date to year, month, hour and day, so we can easily enter those in our algorthms

```python
test_df['count'] = np.nan
train_df['year'] = train_df['datetime'].dt.year
train_df['month'] = train_df['datetime'].dt.month
train_df['hour'] = train_df['datetime'].dt.hour
train_df['day'] = train_df['datetime'].dt.dayofweek

test_df['year'] = test_df['datetime'].dt.year
test_df['month'] = test_df['datetime'].dt.month
test_df['hour'] = test_df['datetime'].dt.hour
test_df['day'] = test_df['datetime'].dt.dayofweek
```

From our decisions as explained above after adding and removing features, we have selected workingday, temp, year, month, hour and day for predictions and we will select x axis and y-axis respectively. x-axis are the features selected and y-axis is the count that we are looking to predict

```python
selected_features =['workingday', 'hour', 'temp', 'humidity', 'windspeed']

#select x, y axis for train data
x_original_train = train_df[selected_features]
y_original_train = train_df['count']

#select x, y axis for test data
x_test = test_df[selected_features]
y_test = test_df['count']
```

Since dataset could be huge and we are not sure which model is best in our situation, lets first create portion of entire set for validations, which will give us the best model options. Here we are going to choose 25% of entrie dataset for validation and random_state is selected at 300.

```python
#validating tests on portion of entire set of data we have
x_train, x_valid, y_train, y_valid = train_test_split(x_original_train, y_original_train, test_size = 0.25, random_state = 300)
x_train.head()
```

```python
x_valid.head()
```

Now as we are provided with the RMSLE (Root Mean Squared Logarithmic Error) in the Kaggle, we will define this as a function:

```python
def RMSLE(predictions, realizations):
    predictions_use = predictions.clip(0)
    rmsle = np.sqrt(np.mean(np.array(np.log(predictions_use+1)-np.log(realizations+1))**2))
    return rmsle
```

Finally its time to check our first model - <b>Regression Tree</b> and we are going to use DecisionTreeRegressor with min_samples_split of 25 and random_state 300 and display the feature importance with this model

```python
regression_tree =DecisionTreeRegressor(min_samples_split = 25, random_state = 300)
regression_tree_model = regression_tree.fit(x_train, y_train)
regression_tree_prediction = regression_tree_model.predict(x_valid)

#checking which feature is important with this model.
pd.DataFrame(regression_tree_model.feature_importances_, index=selected_features)
```

Result shows that hour feature is important as compared to other features. Now we can actually plot the Predicted Vs Actual count using Regress Tree model as below:

```python
plt.figure(figsize=(6,6))
plt.scatter(regression_tree_prediction, y_valid, s=0.2)
plt.xlim(-100, 1200)
plt.ylim(-100, 1200)
plt.plot([-100, 1200], [-100, 1200], color='r', linestyle='-', linewidth=2)
plt.suptitle('Predicted Vs Actual count - Regression Tree', fontsize=20)
plt.xlabel('Regression Tree Prediction', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.show()
```

Finally we will use the RMSLE function to evaluation this model.

```python
RMSLE(regression_tree_prediction, y_valid)
```

Since this is our first model, we don't have anything to compare with, so lets compare this with a second model - <b>Random Forest</b> and we can decide which one is better. Note: Lower RMSLE value is better.

```python
random_forest = RandomForestRegressor(n_estimators = 500, max_features=4, min_samples_leaf=5, random_state = 300)
random_forest_model = random_forest.fit(x_train, y_train)
random_forest_prediction = random_forest_model.predict(x_valid)

pd.DataFrame(random_forest_model.feature_importances_, index=selected_features)
```

This model also indicates that hour is the important feature in this modeling too like in Regression Tree model. We can show this modeling in plot too to compare predicted vs actual count.

```python
plt.figure(figsize=(6,6))
plt.scatter(random_forest_prediction, y_valid, s=0.2)
plt.xlim(-100, 1200)
plt.ylim(-100, 1200)
plt.plot([-100, 1200], [-100, 1200], color='r', linestyle='-', linewidth=2)
plt.suptitle('Predicted Vs Actual count - Random Forest', fontsize=20)
plt.xlabel('Random Forest Prediction', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.show()
```

Lets now check the RMSLE value for this model.

```python
RMSLE(random_forest_prediction, y_valid)
```

Even though difference is not big in our scenarios, still the RMSLE value for Random Forest modeling is lower compared to Regression Tree, so we choose this modeling for our final submission. Now we finally use the Random Forest Modeling to our actual test dataset and predict the total count. Please also note that we have rounded this number to closest integer.

```python
random_forest_model_train = random_forest.fit(x_original_train, y_original_train)
random_forest_model_prediction_test = random_forest_model_train.predict(x_test)
random_forest_prediction_test_clipped = pd.Series(np.round(random_forest_model_prediction_test.clip(0)))
```

The predicted count is stored as Series so far at this stage and now we need to write the series to a csv file in the submission format we are provided. We will use the submissionsample.csv file and create a finalSubmission.csv file.

```python
sample_submission = pd.read_csv("../term_project/sampleSubmission.csv")
sample_submission.head()
random_forest_submission = sample_submission
random_forest_submission['count'] = random_forest_prediction_test_clipped
random_forest_submission.to_csv("../term_project/finalSubmission.csv", index=False)
```
