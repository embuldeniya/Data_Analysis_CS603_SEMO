1) For the Title column, only keep the title part, remove all of the other parts. If the title is inside a
bracket, remove the bracket which encloses the title. If there are multiple …, only keep the
words before the first one. If there are multiple periods (.), only keep words before the first one. 





```python
import numpy as np
import pandas as pd

df = pd.read_csv("BL-Flickr-Images-Book.csv")

def clean_title(itm):
    if itm[0] == '[' and '.' in itm:
        return itm[1:itm.find('.')]
    elif itm[0] == '[':
        return itm[1:itm.find(']')]
    elif '.' in itm:
        return itm[:itm.find('.')]

df_title = pd.DataFrame(df, columns=['Title'])
df_title = df_title.applymap(clean_title)
df["Title"] = df_title    
    
df.head()
```

2) For the Author and Contributors columns, only keep the first author, remove all of the other
parts. If the author or the contributor has other auxiliary information, remove them. title is
inside a bracket, remove the bracket which encloses the title. If there are multiple …, only keep
the words before the first one. If there are multiple periods (.), only keep words before the first
one. For all of the names, they should only have the first letter of the first, middle and last names
be capital letter, all of the remaining letters should be small case. 

```python
def clean_author(itm):
    if '-' in itm:
        return itm[:itm.find('-')]
    elif '|' in itm:
        return itm[:itm.find('|')]
    else:
        return itm

#author
df["Author"] = df["Author"].str.title()
df_author = pd.DataFrame(df, columns = ['Author'])
df_author = df_author.fillna("")
df_author = df_author.applymap(clean_author)
df["Author"] = df_author

#contributors
df["Contributors"] = df["Contributors"].str.title()
df_contrib = pd.DataFrame(df, columns = ['Contributors'])
df_contrib = df_contrib.fillna("")
df_contrib = df_contrib.applymap(clean_author)
df["Contributors"] = df_contrib

df.head()
    
```

3) After cleaning the above columns, save the cleaned data back into a new .csv file. 

```python
df.to_csv('CLEANED-BL-Flickr-Images-Book.csv')
```

4) You may use previous unit learned concepts on multi-level indexing, pivoting, stacking to rearrange the original .csv file to convert summer and winter as secondary level index within the country index, and then generate a new .csv file which will have country as the primary index, then summer and winter as secondary index, and columns of gold, silver and bronze medals.

```python
countries = []

for x in df["Country"][:]:
    if " (" in x:
        x = x[:x.index(" (")]
   
    countries.append(x)
    countries.append(x)
season = []

for i in range(int(len(countries)/2)):
    season.append("Summer")
    season.append("Winter")
    

tuples = list(zip(*[countries, season]))

index = pd.MultiIndex.from_tuples(tuples, names=["Country", "Season"])


data_summer = df.iloc[:,2:6]
data_winter = df.iloc[:,7:11]


data_winter.columns = list(data_summer.columns.values)

stacked = pd.DataFrame()

for i in range(len(data_summer)):
    stacked = stacked.append(data_summer[i:i+1], ignore_index=True)
    stacked = stacked.append(data_winter[i:i+1], ignore_index=True)
    

final = pd.DataFrame(data=stacked.iloc[:,:].values, columns = list(data_summer.columns.values), index=index)

final
```
