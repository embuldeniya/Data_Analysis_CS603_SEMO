```python

for s in test_df:
    s.loc[s["temp"]<= 15, "temp"]= 0
    s.loc[(s["temp"]> 15 & s["temp"]<=20 ),"temp"]= 1
    s.loc[(s["temp"]> 20 & s["temp"]<=20 ),"temp"]= 1
    


```

```python
print (type(test_df["temp"]))
```

0:Cold: <15c (<59f)
Cool: 15-20c (59-68f)
Comfortable: 20-30c (68-86f)
Warm: 30-38c (86-100f)
Hot: >38c (>100f)


```python
def get_tempband(value):
    tempband = None
    if value <=15:
        tempband = 0
    elif value >15 and value <=20:
        tempband = 1
    elif value >20 and value <=30:
        tempband = 2
    elif value >30 and value <=38:
        tempband = 3
    elif value >38:
        tempband = 4
    
    return tempband

test_df["tempband"] = test_df.temp.apply(get_tempband)
```
