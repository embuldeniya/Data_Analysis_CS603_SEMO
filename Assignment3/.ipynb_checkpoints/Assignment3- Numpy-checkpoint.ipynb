```python
profit = []
for i in range (1950,2019):
    print (i)
    profit.append(i)
    t2 = np.where(np.logical_and(date>= np.datetime64(datetime(i,1,1)),date < np.datetime64(datetime(i+1,1,1))))
    v = monthlyAvg(t2,i)
    profit.append()
    profit.append(v)
    
    
val = np.reshape(profit, (-1,2))

print("Most profitable year from 1950 to 2018 is {0}".format(val[(np.argmax(val[:,1])),0].astype(int)))
#print (val[(np.argmax(val[:,1])),0].astype(int))


#return max value from total summation through gain/loss 
#print(int(np.amax(val [:,1], axis=0)))
```
