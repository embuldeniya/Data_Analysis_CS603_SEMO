```python
#list: much the same thing as an array in Java or C or any other language. 
#It’s an ordered collection of objects. You create a listd by enclosing 
#a commaseparated list of elements in square brackets
#automatically grows or shrinks in size as needed
#Python lists can contain different types of elements; a list element can be any Python object.
#Python indexing is more flexible than C indexing; if indices are negative numbers, 
#they indicate positions counting from the end of the list, with –1 being the last
#position in the list, –2 being the second-to-last position, and so forth
x = [2, "two", [1, 2, 3]]
print(x)
print(len(x))
print(x[1])
print(x[-1])
#slicing: list[index1:index2] to extract all items including index1 and up to (but not including) index2 into a new list.
print("slicing: ", x[0:-1])
x[0] = 100
print(x)
#use slice assignment to append a list at back or at front, or remove part of a list
x[len(x):]=['a', 'b', 'c']
print("use slice to append at back: ", x)
x[:0] = ['A', 'B', 'C']
print("use slice to append at front: ", x)
x[6:-1]=[]
print("use slice to remove: ", x)
#append: add a single item at the back of a list
x.append('b')
print(x)
#extend: add a list to the back of a list
x.extend([100,200,300])
print(x)
#insert method is a method of list
x.insert(1, 1000)
print(x)
#del: a general method, can remove a single item or multiple items
del x[1]
print(x)
del x[:2]
print(x)
#remove: find item then delete, not using index
x.remove('C')
print(x)
#reverse
x.reverse()
print(x)
y = [3, 8, 7, 0, 2, 1]
#sort()
y.sort()
print(y)
def compare_num_of_chars(string1):
    return len(string1)
word_list = ['Python', 'is', 'better', 'than', 'C']
#sort() by default sorting lexicography
word_list.sort()
print("default sort: ", word_list)
#can customer sort by definition, but slower than the default one
word_list.sort(key=compare_num_of_chars)
print("sort by definition: ", word_list)
#sort in descending order, set reverse to True
word_list.sort(key=compare_num_of_chars, reverse=True)
print("sort by defintion and in descending order: ", word_list)
z=(3, 8, 7, 0, 2, 1)
#sorted method returns a sorted list for a group of items
z1=sorted(z)
print(z1)
#in: check membership; not in
print("check item's membership in a list:", 3 in z1, 3 not in z1)
#list concatenation
z2=y+z1
print(z2)
#list repitition
z3=y*3
print(z3)
#max, min
print(max(y))
#search returns index
print(y.index(7))
y.append(8)
#find num of occurences of an item
print(y.count(8))
```

```python
#use nested list for multi-dimensional list
m = [[0, 1, 2], [10, 11, 12], [20, 21, 22]]
print(m[0])
print(m[0][1])

#list is mutuble, so be careful of shallow copy vs deep copy
nested = [0,1]
original =[nested, 100, 200, 300]
print(original)
#shallow copy
nested[0] = -100
print(original)
original[0][1] = -200
print(nested)
#reset nested to a new list, then original won't be impacted
nested = [51, 52, 53]
print(original)
print(nested)
#using slicing is shallow copy
shallow = original[:]
import copy
#deepcopy() is for deep copy
deep = copy.deepcopy(original)
shallow[0][0]="Zero"
print(original)
deep[0][0]="deep zero"
print(original)
print(deep)
```

```python
"""
Using extend to append  elements to an existing list, especially if you are building up a large list, 
is usually preferable, is faster than the concatenative alternative
"""
x1 = [4, None, 'foo']
x2 = [7, 8, (2, 3)]
print(x1 + x2)
x1.extend(x2)
print(x1)
```

```python
"""
built-in function, enumerate, which returns a sequence of (i, value) tuples
"""
some_list = ['foo', 'bar', 'baz']
mapping = {}
for i, v in enumerate(some_list):   
    mapping[v] = i
print(mapping)
"""
zip “pairs” up the elements of a number of lists, tuples, or other sequences to create a list of tuples
"""
seq1 = ['foo', 'bar', 'baz']
seq2 = ['one', 'two', 'three']
zipped = zip(seq1, seq2)
print(list(zipped))
seq3 = [False, True]
"""
zip can take an arbitrary number of sequences, and the number of elements it produces is        
determined by the shortest sequence
"""
print(list(zip(seq1, seq2, seq3)))
"""
A very common use of zip is simultaneously iterating over multiple sequences, 
possibly also combined with enumerate
"""
for i, (a, b) in enumerate(zip(seq1, seq2)):
    print('{0}: {1}, {2}'.format(i, a, b))
"""
zip can be applied in a clever way to  “unzip” the sequence. 
converting a list of rows into a list of columns
"""
pitchers = [('Nolan', 'Ryan'), ('Roger', 'Clemens'), ('Schilling', 'Curt')]
first_names, last_names = zip(*pitchers)
print(first_names)
print(last_names)
```
