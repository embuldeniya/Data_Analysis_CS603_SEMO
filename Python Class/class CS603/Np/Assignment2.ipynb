```python
import numpy as np 
from datetime import datetime

file = "/Users/techyouknow/Desktop/Python Class/SP500.csv"


date =np.loadtxt(file,  delimiter = "," , skiprows=1, usecols=(0) , dtype=np.datetime64)

value=np.loadtxt(file,  delimiter = "," , skiprows=1, usecols=(1,2,3,4,5,6) , dtype=np.float64)


#No1 daily gain from open & close  
gain = [i[0] - i[3] for i in value]

print("** highest daily gain value  & date **")
print(date[np.argmax(gain, axis=0)])
print(np.amax(gain, axis=0))


print("** lowest daily gain value & date **")
print(date[np.argmin(gain, axis=0)])
print(np.amin(gain, axis=0))


#No2 volume & date
print("** daily volume transcation and date  **")
print(int(np.amax(value, axis=0)[5]))
print(date[(np.argmax(value, axis=0)[5])])

print("")

#No3 monthly Avg & open price range in 2017/2018  

def monthlyAvg(t1,x):
   
    s= np.array (t1)
    '''
    for x in t1:
        s = x
    
    
    '''
    mon = []    
    for i in s:
        mon.append(value [i]) 
    
    mon = np.array (mon)

    print ("{0} - {1} \n".format(date[s[0]],date[s[-1]]))
  
    
    print ("open average= ",np.mean(mon[:,0] ))
    print ("high average= ",np.mean(mon[:,1] ))
    print ("low average= ",np.mean(mon[:,2] ))
    print ("close average= ",np.mean(mon[:,3] ))
    print ("adjclose average= ",np.mean(mon[:,4] ))
    print ("volume average= ",np.mean(mon[:,5] ))  
    print ("\n === Open Price Range === ")
    print ("{0} -  {1}".format(np.min(mon[:,0]),np.max(mon[:,0])))
    print ("\n")


month = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug","Sep", "Oct", "Nov","Dec"]  

print ("Months between  2017/2018 \n")
for i in range (1,13):
    if (i == 12):
        print (month [i-1])
        t1 = np.where(np.logical_and(date> np.datetime64(datetime(2017,i,1)),date< np.datetime64(datetime(2018,1,1))))
        monthlyAvg(t1,i)
    else:
        print (month [i-1])
       
        t1 = np.where(np.logical_and(date> np.datetime64(datetime(2017,i,1)),date< np.datetime64(datetime(2017,i+1,1))))
        monthlyAvg(t1,i)

        
```
